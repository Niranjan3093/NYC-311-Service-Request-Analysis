## 1. Data Preparation

### 1.1. Import the dataset

Code:

```python
# Import the pandas  library for data manipulation
import pandas as pd  

# Load the dataset from a CSV file with memory optimization disabled
df = pd.read_csv('Customer Service_Requests_from_2010_to_Present.csv', low_memory=False)

print("Dataset imported successfully.")

# Display the first five rows of the dataset to preview its structure
df.head()
```

### 1.2. Provide your insight on the information and details that the provided dataset carries.

Code:

```python
# Display column names, non-null counts, and data types
print("Dataset Info:")
print(df.info())

# Show descriptive statistics for all columns, including categorical
print("\nDescriptive Statistics:")
df.describe(include='all')
```

### 1.3. Convert the columns "Created Date" and "Closed Date" to datetime datatype and create a new column "Request_Closing_Time" as the time elapsed between request creation and request closing.

Code:

```python
# Converting the columns "Created Date" and "Closed Date" to datetime datatype
# - errors='coerce' turns any unparsable value into NaT (missing datetime).
df['Created Date'] = pd.to_datetime(df['Created Date'], errors='coerce')

df['Closed Date'] = pd.to_datetime(df['Closed Date'],  errors='coerce')
    
# Creating a new column "Request_Closing_Time" 
df['Request_Closing_Time'] = df['Closed Date'] - df['Created Date']

print("Request_Closing_Time column created.")

df[['Created Date', 'Closed Date', 'Request_Closing_Time']].head()
```

### 1.4. Drop the irrelevant columns

Code:

```python
# Define a list of columns to drop.
columns_to_drop = [
    'Agency Name', 'Incident Address', 'Street Name', 'Cross Street 1', 'Cross Street 2',
    'Intersection Street 1', 'Intersection Street 2', 'Address Type', 'Park Facility Name', 
    'Park Borough', 'School Name', 'School Number', 'School Region', 'School Code',
    'School Phone Number', 'School Address', 'School City', 'School State', 'School Zip',
    'School Not Found', 'School or Citywide Complaint', 'Vehicle Type', 'Taxi Company Borough',
    'Taxi Pick Up Location', 'Bridge Highway Name', 'Bridge Highway Direction', 'Road Ramp',
    'Bridge Highway Segment', 'Garage Lot Name', 'Ferry Direction', 'Ferry Terminal Name',
    'Landmark', 'X Coordinate (State Plane)', 'Y Coordinate (State Plane)', 'Due Date',
    'Resolution Action Updated Date', 'Community Board', 'Facility Type', 'Location'
]

# Drop the irrelevant columns
df.drop(columns=columns_to_drop, axis=1, inplace=True, errors='ignore')
print("Irrelevant columns dropped.")
df.head()
```

### 1.5. Write a python program to remove the NaN missing values from updated dataframe. 

Code:

```python
# Remove all rows that contain any missing (NaN) values in any column.
df_cleaned = df.dropna()

# Print how many rows were removed during cleaning.
print("Missing values handled. Number of rows before cleaning: {}, after cleaning: {}"
      .format(len(df), len(df_cleaned)))
```

## 4. Statistical Testing

### 4.1. Test 1:  Whether the average response time across complaint types is similar or not.

#### ● State the Null Hypothesis (H0) and Alternate Hypothesis (H1).
#### ● Perform the statistical test and provide the p-value.
#### ● Interpret the results to accept or reject the Null Hypothesis.


Code:

```python
from scipy.stats import f_oneway

# Convert the "Request_Closing_Time" timedelta column into hours (float) for analysis
df_cleaned.loc[:, 'Request_Closing_Hours'] = (
    df_cleaned['Request_Closing_Time'].dt.total_seconds() / 3600
)

# Identify the top 3 most frequent complaint types for comparison
top3 = df_cleaned['Complaint Type'].value_counts().nlargest(3).index.tolist()

# Extract the closing time data (in hours) for each of the top 3 complaint types
# Each group's data will be used as a sample in the ANOVA test
groups = [
    df_cleaned.loc[df_cleaned['Complaint Type'] == c, 'Request_Closing_Hours']
    for c in top3
]

# Perform one-way ANOVA to test if the average response times differ across complaint types
F_stat, p_val = f_oneway(*groups)

# Output the F-statistic and p-value to interpret the test result
print(f"ANOVA result: F = {F_stat:.3f}, p = {p_val:.3f}")
```

### 4.2. Test 2: Whether the type of complaint or service requested and location are related.

#### ● State the Null Hypothesis (H0) and Alternate Hypothesis (H1).
#### ● Perform the statistical test and provide the p-value.
#### ● Interpret the results to accept or reject the Null Hypothesis.


Code:

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Create a contingency table showing the frequency of each complaint type across boroughs
# Rows: Complaint Types | Columns: Boroughs | Values: Count of occurrences
contingency = pd.crosstab(df_cleaned['Complaint Type'], df_cleaned['Borough'])

# Perform the chi-square test of independence to determine if complaint type is related to borough
chi2, p_val, dof, expected = chi2_contingency(contingency)

# Print the chi-square statistic, p-value, and degrees of freedom
print(f"Chi-square result: χ² = {chi2:.2f}, p = {p_val:.3f}, dof = {dof}")
```
